# Desafios da aula 01 do Segundo bootcamp de data science da Alura

```python
import pandas as pd

data_url = 'https://github.com/alura-cursos/agendamento-hospitalar/raw/main/dados/A151346189_28_143_208.csv'

data = pd.read_csv(data_url, encoding="ISO-8859-1", 
                    skiprows = 3, sep=";", skipfooter=12,
                    thousands=".", decimal=",")
```

```python
pd.options.display.float_format = "{:.2f}".format

data.head()
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

ax = data.plot(x="Unidade da Federação", y="2008/Ago", kind="bar", figsize=(9,6))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))

plt.title("Valor por unidade da federação")
plt.show()
```

## Desafio 01

No desafio 01 é solicitado que o gráfico acima fosse renomeado de forma que passa melhor a informação que queremos. No caso, um título melhor seria:
**Valor gasto em reais no SUS por unidade federativa**, pois sintetiza melhor o que estamos vendo de forma sucinta.

```python
ax = data.plot(x="Unidade da Federação", y="2008/Ago", kind="bar", figsize=(12,8))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.title("Valor gasto em reais no SUS por unidade federativa")
plt.show()
```

## Desafio 02

O segundo é solicitado apenas para realizarmos a mesma analise, mas para o mês mais recente. No caso, maio/2021. Contudo, nos dados disponibilizados pela alura, só vão até 2020/ago, e, no site do Datasus, só temos até março/2021. Dessa forma, é necessário obter a base completa para obtermos os valores de **março/2021** no site do **Datasus**. 

```python
mar_2021_valor_total_UF = pd.read_csv('..\dados\mar_2021_valor_total_UF.csv', encoding="ISO-8859-1", 
                    skiprows = 3, sep=";", skipfooter=12,
                    thousands=".", decimal=",")

mar_2021_valor_total_UF
                    
```

```python
mar_2021_valor_total_UF['Unidade da Federação'] = mar_2021_valor_total_UF['Unidade da Federação'].apply(lambda x: ''.join(x.split(' ')[1:]))

ax = mar_2021_valor_total_UF.plot(x="Unidade da Federação", y="2021/Mar", kind="bar", figsize=(12,8))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.0f}"))
plt.title("Valor gasto em reais no SUS por unidade federativa")
plt.show()
```

No codigo acima, só foi repetido o mesmo plot de barras que tinha sido apresentado previamente, mas pro mes de março de 2021. Foi feita uma pequena alteração na coluna das "Unidades de Federação" para retirar o numero identificador delas, que só prejudicavam na identificação.
